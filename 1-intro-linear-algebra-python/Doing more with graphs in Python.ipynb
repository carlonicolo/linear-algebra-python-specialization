```python
#pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Basic Example

```python
x = np.linspace(-3,3,20)

y = x**2
x
```

```python
plt.plot(x,y,'b-')
```

### Basic Matplotlib Commands

```python
plt.plot(x,y, 'g*-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Our first Plot')
```

### Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')
plt.title('Y is function of X')

plt.subplot(1,2,2)
plt.plot(y,x, 'r--')
plt.title('X is a function of Y')
```

### Introduction to the Object Oriented Method

```python
fig = plt.figure(figsize=(10,8))
axe1 = fig.add_axes([0.1,0.1,0.8,0.8])
axe1.plot(x,y,'b-o',lw=2)
axe1.set_title('Larger Graph')

axe2 = fig.add_axes([0.3,0.1,0.2,0.3])
axe2.plot(y,x,'r->',lw=2)
axe2.set_title('Saller Graph')
```

### Subplots()  
The subplots() object will act as a more automatic axis manager

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

Then you can specify the number of rows and columns when creating the subplots() object:

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
```

```python
# Axes is an array of axes to plot on
axes
```

We can iterate throught this array

```python
for ax in axes:
    ax.plot(x,y, 'k')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title('Repeating')
    
# display the figure object
fig
```

### Legends, labels and titles  
Now that we have covered the basics of how to create a figure and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.  
  
#### Figure titles  
A little can be added to each axis instance in a figure. To set the title, use the **set_title** method in the axes instance:

```python
ax.set_title("title");
```

```python
# Axis labels
ax.set_xlabel("x")
ax.set_ylabel("y")
```

### Legends
You can use the label="label text" keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, 'r*', label="x**2")
ax.plot(x, x**3, 'g--', label="x**3")
ax.legend(loc=0)

```

```python
# Legend options...
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# ... may more options are available

# Most common choice
ax.legend(loc=4) # let matplotlib decide the optimal location
fig
```
